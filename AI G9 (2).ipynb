```python
import pandas as pd

# Group the crime report data by 'REPORT_YEAR' and count the number of cases in each year
case_by_year = cm.groupby('REPORT_YEAR').size().reset_index(name='total_cases')

# Convert case_by_year.values to a list
tc = pd.DataFrame({'Year': case_by_year['REPORT_YEAR'].values.tolist(), 'total_cases': case_by_year['total_cases'].values.tolist()})

data = {
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Population': [5797000, 5867000, 5938000, 6010000, 6082000, 6139000, 6197000, 6255000, 6313000]
}


pop = pd.DataFrame(data)

merged_data = pd.merge(tc, pop, left_on='Year', right_on='Year')

# Calculate the crime rate by dividing the number of cases by the population and multiplying by 100,000
merged_data['Crime_Rate'] = (merged_data['total_cases'] / merged_data['Population']) * 100000

print(merged_data[['Year', 'Crime_Rate']])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Input data
data = {
    'Death_Injuries': [103, 152, 193, 187, 214, 284, 216, 209],
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'Population': [5797000, 5867000, 5938000, 6010000, 6082000, 6139000, 6197000, 6255000],
    'Crime rate': [550.526134, 554.218510, 555.557427, 574.492512, 603.946070, 637.740039, 557.753752, 551.446843],
    'CSI': [56.28, 56.90, 59.56, 61.26, 66.83, 68.21, 57.80, 56.71]
}

df = pd.DataFrame(data)

# Create the feature matrix X
X = df[['Population', 'Crime rate', 'CSI']]

# Create the target vector y
y = df['Death_Injuries']

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Generate points within the range of the independent variables
population_vals = np.linspace(df['Population'].min(), df['Population'].max(), 50)
crime_rate_vals = np.linspace(df['Crime rate'].min(), df['Crime rate'].max(), 50)
csi_vals = np.linspace(df['CSI'].min(), df['CSI'].max(), 50)
population_mesh, crime_rate_mesh, csi_mesh = np.meshgrid(population_vals, crime_rate_vals, csi_vals)
X_surface = np.column_stack((population_mesh.ravel(), crime_rate_mesh.ravel(), csi_mesh.ravel()))

# Predict the response variable for the regression surface
y_surface = model.predict(X_surface)

# Create a larger 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the regression surface
ax.scatter(df['Population'], df['Crime rate'], df['CSI'], c=df['Death_Injuries'], cmap='viridis', label='Data')
ax.scatter(population_mesh.ravel(), crime_rate_mesh.ravel(), csi_mesh.ravel(), c=y_surface, cmap='plasma', alpha=0.2, label='Regression Surface')

# Add 3D points
ax.scatter(df['Population'], df['Crime rate'], df['CSI'], c=df['Death_Injuries'], cmap='viridis', label='Data Points')

# Set labels and title
ax.set_xlabel('Population')
ax.set_ylabel('Crime rate')
ax.set_zlabel('CSI')
ax.set_title('Multiple Linear Regression: Regression Surface')

# Add a colorbar
cbar = fig.colorbar(ax.scatter(df['Population'], df['Crime rate'], df['CSI'], c=df['Death_Injuries'], cmap='viridis'))
cbar.ax.set_ylabel('Death_Injuries')

# Adjust the aspect ratio of the plot
ax.dist = 12

# Show the plot
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

# Input data
data = {
    'Death_Injuries': [103, 152, 193, 187, 214, 284, 216, 209],
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'Population': [5797000, 5867000, 5938000, 6010000, 6082000, 6139000, 6197000, 6255000],
    'Crime rate': [550.526134, 554.218510, 555.557427, 574.492512, 603.946070, 637.740039, 557.753752, 551.446843],
    'CSI': [56.28, 56.90, 59.56, 61.26, 66.83, 68.21, 57.80, 56.71]
}

df = pd.DataFrame(data)

# Create the design matrix
X = sm.add_constant(df[['Population', 'Crime rate', 'CSI']])

# Create the response variable
y = df['Death_Injuries']

# Create an OLS model
model = sm.OLS(y, X)

# Fit the model
results = model.fit()

# Print the model summary
print(results.summary())
```
