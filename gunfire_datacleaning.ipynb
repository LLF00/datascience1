```python
df2 = df[df["HOOD_158"].str.contains("NSA")==False]
df2 = df[df["NEIGHBOURHOOD_158"].str.contains("NSA")==False]
df2 = df[df["HOOD_140"].str.contains('NSA')==False]
df2 = df[df['NEIGHBOURHOOD_140'].str.contains('NSA')==False]
print(df2)
```

It shows missing values in HOOD&NEIGHBORHOOD also with missing value on geographic coordinates. So # of rows does not change. There are 5641 rows are valid.

```python
df3 = df2[df2.X !=0]
df3 = df3[df3.Y !=0]
df3
```

```python
conditions = [(df3['DEATH'] >= 1) & (df3['INJURIES']>=0), (df3['DEATH'] <1) & (df3['INJURIES']>=1), (df3["DEATH"]<1) & (df3["INJURIES"]<1)]
choices = ["Red","Yellow","Green"] 

df3["Classification"] = np.select(conditions, choices, default=0)
```
