```python
import matplotlib.pyplot as plt

# Data
banks = ['JPMChase', 'BOA', 'Citi', 'WellsFargo', 'TD', 'CapitalOne', 'Average']
years = [2021, 2022, 2023]
scores = [
    [828, 849, 686],
    [823, 834, 689],
    [797, 817, 656],
    [803, 821, 653],
    [815, 831, 670],
    [804, 804, 689],
    [810, 815, 652]
]

# Plotting
plt.figure(figsize=(10, 6)) 
for bank, score in zip(banks, scores):
    plt.plot(years, score, marker='o', label=bank)

plt.xlabel('Year')
plt.ylabel('Score')
plt.title('Scores for Each Bank (2021-2023)')
plt.legend()
plt.grid(True)
plt.xticks(years)  
plt.tight_layout()  
plt.show()

```

###
1. The overall economic trend is not good. That has led to a deterioration in the financial situation of some customers, such as declining stock prices and inflation. In this scenario, only about 21% of people received advice and guidance from banks.

2. There is an increased risk of encountering fraud.
