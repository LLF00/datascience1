```python
import matplotlib.pyplot as plt

df['Release Time'] = pd.to_datetime(df['Release Time'], format='%m/%d/%Y', errors='coerce')

# Extract the year from the 'Release Time' column
df['Release Year'] = df['Release Time'].dt.year

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Release Year'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Release Year Distribution')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Retweet'], bins=20, color='skyblue', edgecolor='black')
plt.title('Retweet Distribution')
plt.xlabel('Number of Retweets')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Like'], bins=20, color='skyblue', edgecolor='black')
plt.title('Retweet Distribution')
plt.xlabel('Number of Retweets')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

#Creating the text variable
text = " ".join(post for post in df.Post)

#We can generate the wordcloud in the following manner.
# Generate word cloud
word_cloud = WordCloud(
    width=6000,
    height=4000,
    random_state=1,
    background_color="black",
    colormap="Pastel1",
    collocations=False,
    stopwords=STOPWORDS,
    ).generate(text)


# Display the generated Word Cloud
plt.imshow(word_cloud)
plt.axis("off")
plt.savefig(r'D:\5.ACP\cloudsample.png')

plt.show()
```
