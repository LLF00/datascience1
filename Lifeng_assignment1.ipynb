Use dropna() to drop any rows with null value permentally.
Assign effective rows to df2, use inplace=False.
If inplace= True, then they are dataframe with null values in rows.
axis=0, drop rows with null values. axis=1, drop columns with null values.
how='any', drop the row/column if any value is null.
how='all', drop the row/column if all values are null.

```python
df2 = df.dropna(axis=0, how="any", inplace=False)
```

To create pie chart for regions. Use region.unique() to count.

```python
df2.Region.unique()
```

For Region 2, Product 26, 16 & 2 are top 3 bestsellers.

```python
region2 = df2.loc[df2['Region'] == 'Region_2']
agg_functions = {'Sale Revenues':'sum'}
region2_combine = region2.groupby(region2['Product']).aggregate(agg_functions)
region2_combine = region2_combine.sort_values('Sale Revenues', ascending=False)
data = region2_combine["Sale Revenues"]
labels = region2_combine.index
plt.pie(data, labels=labels)
plt.show
```
