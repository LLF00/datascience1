```python
df3 = df2.loc[df2['brand'].isin(['samsung', 'apple']), ['brand', 'price']]
df3
```

assumption :$500 - $1000 apple sells more products than samsung

```python
from scipy import stats
```

```python
# Convert 'price' to a categorical variable
df3['price_table'] = pd.cut(df3['price'], bins=[0, 300, 600, 900, 1200, 1500])

# Create a contingency table
contingency_table = pd.crosstab(df3['brand'], df3['price_table'])

chi2, p_val, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-squared: {chi2}")
print(f"P-value: {p_val}")
```
